```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_excel("vaccine_data.xlsx")
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()   #extra space
df.columns = df.columns.str.lower()   #small letters

#rename columns name in simple name
df.columns = ["userid", "name", "first_status","first_date", "second_status","second_date"]

print(df.columns)
```

```python
df["first_date"] = pd.to_datetime(df["first_date"], errors="coerce")
df["second_date"] = pd.to_datetime(df["second_date"], errors="coerce")
```

```python
print(df.head())
print(df.columns)
```

```python
user_id = input("Enter User ID (example p001): ")
user = df[df["userid"] == user_id]
```

```python
if user.empty:
    print("User not found. Please check and try again.")

if not user.empty:

    name = user.iloc[0]["name"]
    first_status = user.iloc[0]["first_status"]
    second_status = user.iloc[0]["second_status"]
    first_date = user.iloc[0]["first_date"]

    today = datetime.today()

    # case 1: Both doses taken
    if first_status == "yes" and second_status == "yes":
        print("Hi", name + ", Congratulations! You are fully vaccinated.")

    # Case 2: No dose taken 
    if first_status == "no" and second_status == "no":
        print("Hi", name + ", Please take your first dose at earliest.")

    # Case 3 & 4: Only first dose taken
    if first_status == "yes" and second_status == "no":

        days= (today- first_date).days
        first_date_format = first_date.strftime("%d/%m/%Y")

        next_dose = first_date + pd.Timedelta(days=90)
        next_dose_format = next_dose.strftime("%d/%m/%Y")

        if days > 90:
            print("Hi", name + ", You have taken your first dose on", 
                 first_date_format + ", Please take your 2nd dose at earliest.")

        if days <= 90:
            print("Hi", name + ", You have taken your first dose on",
                 first_date_format + ", please take your 2nd dose on or after",
                 next_dose_format + ".")
```
